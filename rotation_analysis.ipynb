```python
%matplotlib inline
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df1 = pd.read_csv("/Users/haas/Downloads/Rotation1/Crop_Names.csv")
df1["lat"] = [float(x[:6]) for x in df1.Grid]
df1["lon"] = [float(x[7:]) for x in df1.Grid]
newdf1 = df1.set_index(["lat", "lon"])
ds1 = xr.Dataset()
ds1 = ds1.from_dataframe(newdf1)
deds1 = ds1.sel(lat=slice(48.125,53.125), lon=slice(7.125,13.125))
print("Pic some grid cell in Germany ans print the crop sequence for rotation 1.")
deds1.sel(lat=49.125,lon=11.125)
```

```python
df2 = pd.read_csv("/Users/haas/Downloads/Rotation2/Crop_Names_c2.csv")
df2["lat"] = [float(x[:6]) for x in df2.Grid]
df2["lon"] = [float(x[7:]) for x in df2.Grid]
newdf2 = df2.set_index(["lat", "lon"])
ds2 = xr.Dataset()
ds2 = ds2.from_dataframe(newdf2)
deds2 = ds2.sel(lat=slice(48.125,53.125), lon=slice(7.125,13.125))
print("Pic some grid cell in Germany ans print the crop sequence for rotation 2.")
deds2.sel(lat=49.125,lon=11.125)
```

```python
wheat = {}
barley = {}
maize = {}
rape = {}
sugb = {}
pota = {}
print("Analysis of rotation 1 and 2 and the sum of both.")
for y in range(1952,2004):
    print(y)
    field1 = deds1[str(y)].values.flatten()
    field2 = deds2[str(y)].values.flatten()
    field3 = np.concatenate((field1,field2))
    print( field3 )
    c1 = Counter( [x for x in field1.ravel() ] )
    c2 = Counter( [x for x in field2.ravel() ] )
    c3 = Counter( [x for x in field3.ravel() ] )
    print("\tRot 1:", c1, sum(c1.values() ))
    print("\tRot 2:", c2, sum(c2.values() ))
    print("\tRot 1+2:", c3, sum(c3.values() ))
    
    f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharex='all', sharey='all',  figsize=(10,4))
    ax1.hist( field1 )
    ax2.hist( field2 )
    ax3.hist( field3 )
    plt.show()
    wheat[y] = c3["WSWH"]
    barley[y] = c3["SBAR"]
    maize[y] = c3["MAIF"]
    rape[y] = c3["RAPE"]
    sugb[y] = c3["SUGB"]
    pota[y] = c3["POTA"]
    
```

```python
print("Summary of the comination of rotation 1 + rotation2!!! so we have in total in this statistic twice the crops as grid cells.")
plt.title("Rotation c1 + c2 major crops")
plt.plot(*zip(*sorted(wheat.items() )), label="total wheat")
plt.plot(*zip(*sorted(barley.items() )), label="total barley")
plt.plot(*zip(*sorted(maize.items() )), label="total silage maize")
plt.plot(*zip(*sorted(rape.items() )), label="total rape seeds")
plt.plot(*zip(*sorted(sugb.items() )), label="total sugar beet")
plt.plot(*zip(*sorted(pota.items() )), label="total potato")
plt.legend(ncol=3)
plt.show()
```
